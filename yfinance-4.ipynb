```python
import nest_asyncio
nest_asyncio.apply()

from telegram import Update
from telegram.ext import Application, MessageHandler, filters, ContextTypes
from agno.agent import Agent
from agno.models.groq import Groq
from agno.tools import tool
from agno.tools.mcp import MCPTools
from agno.db.sqlite import SqliteDb
import requests
import asyncio

# Configuration VAPI

TELEGRAM_BOT_TOKEN = "8288920674:AAG9iLGHrS6wAtwiSaCnPib2ZpGn1OYPOWw"

@tool
def trigger_voice_call(message: str, user_phone_number: str) -> str:
    """Trigger a voice call to notify user."""
    url = "https://api.vapi.ai/call"
    headers = {
        "Authorization": f"Bearer {VAPI_API_KEY}",
        "Content-Type": "application/json"
    }
    payload = {
        "assistantId": VAPI_ASSISTANT_ID,
        "phoneNumberId": VAPI_PHONE_NUMBER_ID,
        "customer": {"number": user_phone_number},
        "assistantOverrides": {"variableValues": {"message": message}}
    }
    try:
        response = requests.post(url, headers=headers, json=payload)
        return f"‚úÖ Appel initi√©!" if response.status_code == 201 else f"‚ùå Erreur: {response.text}"
    except Exception as e:
        return f"‚ùå Erreur: {e}"

# Global agent variable
agent = None

async def setup_agent():
    global agent

    # Tavily MCP
    tavily_mcp = MCPTools(
        transport="streamable-http",
        url="https://mcp.tavily.com/mcp/?tavilyApiKey=your-tavily-key"
    )
    await tavily_mcp.connect()

    db = SqliteDb(db_file="telegram_agent.db")

    agent = Agent(
        name="Stock Broker Agent",
        model=Groq(id="openai/gpt-oss-120b"),  # Valid Groq model
        tools=[trigger_voice_call, tavily_mcp],
        description="Tu es un courtier en bourse professionnel.",
        instructions=[
            "Tu peux appeler des clients pour proposer des opportunit√©s d'investissement.",
            "Tu peux rechercher des informations sur le web avec Tavily.",
        ],
        db=db,
        markdown=True,
    )
    return tavily_mcp

async def handle_message(update: Update, context: ContextTypes.DEFAULT_TYPE):
    user_message = update.message.text
    response = await agent.arun(user_message)
    await update.message.reply_text(response.content)

async def main():
    tavily_mcp = await setup_agent()

    app = Application.builder().token(TELEGRAM_BOT_TOKEN).build()
    app.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, handle_message))

    try:
        await app.run_polling()
    finally:
        await tavily_mcp.close()

if __name__ == "__main__":
    asyncio.run(main())
```
